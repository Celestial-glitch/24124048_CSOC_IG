```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
metrics = {
    'Pure Python': {
        'Train': {'MAE': 29030.96018315766, 'RMSE': 46827.182644055145, 'R²':  0.8359649330725991},
        'Test': {'MAE': 29132.204990169415, 'RMSE': 49530.28203648108, 'R²': 0.8127875434807676}
    },
    'NumPy': {
        'Train': {'RMSE': 47437.70531715659,
                  'MAE': 29390.753190582273,
                   'R²': 0.8305390388496584},
        'Test': {'RMSE': 47265.51859560591,
                 'MAE': 30057.245767037537,
                 'R²': 0.8340230136558807}
    },
    'Scikit-learn': {
        'Train': {'MAE': 27835.55,
                  'RMSE': 44849.49,
                  'R²': 0.8495
},
        'Test': {'MAE': 28492.46, 'RMSE': 47868.46, 'R²': 0.8251}
    }
}
```

```python
import matplotlib.pyplot as plt

# Prepare lists for each metric
methods = list(metrics.keys())
metrics_list = ['MAE', 'RMSE', 'R²']

for metric in metrics_list:
    train_values = [metrics[method]['Train'][metric] for method in methods]
    test_values = [metrics[method]['Test'][metric] for method in methods]

    x = range(len(methods))
    width = 0.35

    plt.figure(figsize=(8, 5))
    plt.bar([i - width/2 for i in x], train_values, width=width, label='Train', color='skyblue')
    plt.bar([i + width/2 for i in x], test_values, width=width, label='Test', color='salmon')

    plt.xlabel('Method')
    plt.ylabel(metric)
    plt.title(f'Comparison of {metric} across Methods')
    plt.xticks(x, methods)
    plt.legend()
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
