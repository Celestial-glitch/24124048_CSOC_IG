```python
df = pd.read_csv("housing.csv")
```

Filling missing total_bedrooms with median value

```python
df["total_bedrooms"] = df["total_bedrooms"].fillna(df["total_bedrooms"].median())
```

Converting categorical data to numerical using One-hot encoding for ocean_proximity

```python
print(df['ocean_proximity'].value_counts())  

```

```python
df = pd.get_dummies(df, columns=['ocean_proximity'], drop_first=True)
print(df.columns) # Checking the new columns after encoding
```

```python
df.info()
```

```python
df[['ocean_proximity_INLAND', 'ocean_proximity_ISLAND', 'ocean_proximity_NEAR BAY', 'ocean_proximity_NEAR OCEAN']] = \
    df[['ocean_proximity_INLAND', 'ocean_proximity_ISLAND', 'ocean_proximity_NEAR BAY', 'ocean_proximity_NEAR OCEAN']].astype(float)
```

```python
df.info()
```

```python
df.head()
```

Added Feature: Price-to-Income Ratio


```python
df["price_to_income_ratio"] = df["median_house_value"] / df["median_income"]
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

performing feature rescaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
columns_to_scale = [
    "longitude", "latitude", "housing_median_age",
    "total_rooms", "total_bedrooms", "population",
    "households", "median_income", "price_to_income_ratio"
]
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[columns_to_scale] = scaler.fit_transform(df_scaled[columns_to_scale])
```
